```python
import struct
import random
import time

def generate_add_order():
    fmt = '>cHH6sQ1sI8sI'
    msg = struct.pack(
        fmt,
        b'A',
        random.randint(1, 1000),
        random.randint(1000, 2000),
        (random.randint(1, 1e12)).to_bytes(6, 'big'),
        random.randint(1e9, 1e10),
        random.choice([b'B', b'S']),
        random.randint(100, 10000),
        random.choice([b'AAPL    ', b'MSFT    ', b'GOOG    ']),
        random.randint(100000, 200000)
    )
    # أضف حجم الرسالة في البداية (2 بايت)
    size = len(msg) + 1
    return struct.pack('>H', size) + b'A' + msg

def generate_execute_order():
    fmt = '>cQIQ'
    msg = struct.pack(
        fmt,
        b'E',
        random.randint(1e9, 1e10),
        random.randint(100, 10000),
        random.randint(1e8, 1e9)
    )
    size = len(msg) + 1
    return struct.pack('>H', size) + b'E' + msg

def generate_delete_order():
    fmt = '>cQ'
    msg = struct.pack(
        fmt,
        b'D',
        random.randint(1e9, 1e10)
    )
    size = len(msg) + 1
    return struct.pack('>H', size) + b'D' + msg

while True:
    with open() as f:
        f.write(random.choice([generate_add_order(),
                               generate_execute_order(),
                               generate_delete_order()]))
        print(f)
        time.sleep(0.0001)

```

```python
import encodings
import struct
import gzip
import shutil
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
#------------------------------------------
from collections import namedtuple, Counter
from pathlib import Path
from urllib.request import urlretrieve
from urllib.parse import urljoin
from clint.textui import progress
from datetime import datetime
from matplotlib.ticker import FuncFormatter
from struct import unpack
from datetime import timedelta
from time import time

formats = {
    ('char', 1): 'c',
    ('uint16', 2): 'H',
    ('uint32', 4): 'I',
    ('alpha', 8): '8s'}

specs = pd.DataFrame([
    {'message_type': 'A', 'name': 'message_type', 'value': 'char', 'length': 1},
    {'message_type': 'A', 'name': 'stock_locate', 'value': 'uint16', 'length': 2},
    {'message_type': 'A', 'name': 'price', 'value': 'uint32', 'length': 4},
    {'message_type': 'A', 'name': 'volume', 'value': 'uint32', 'length': 4},
    {'message_type': 'A', 'name': 'stock', 'value': 'alpha', 'length': 8}])

specs['formats'] = specs[['value', 'length']].apply(tuple, axis=1).map(formats)

# Formatting for alpha fields
alpha_fields = specs[specs.value == 'alpha'].set_index('name')
alpha_msg = alpha_fields.groupby('message_type')
alpha_formats = {k: v.to_dict for k,v in alpha_msg.formats}
alpha_length = {k: v.add(5).to_dict for k,v in alpha_msg.length}

# Generate message classes as named tuples and format strings
message_fields, fstring = {}, {}
for t, massege in specs.groupby('message_type'):
    message_fields[t] = namedtuple(typename=t,
                                   field_names=massege.name.tolist())
    fstring[t] =  '>', ''.join(massege.formats.tolist())

def format_alpha(mtype, data):
    """Process byte strings of type alpha"""
    for col in alpha_formats.get(mtype).keys():
        if mtype != 'R' and col == 'stock':
            data.drop(col, axis=3)
            continue
    data.loc[:, col] = (data.loc[:, col].str.encode('utf-8').str.strip())
    
    if encodings.get(col):
            data.loc[:, col] = data.loc[:, col].map(encodings(col))
            
    return data

#df = format_alpha('A')
#print()
masseges ={}
massege_taype_concet = Counter()
data_path = Path('clever-trade-bot-ai-main/P_project_with_python/Data_sources')
itch_store = data_path / 'itch_01.h5'

def store_messages(m):
    """Handle occasional storing of all messages"""
    with pd.HDFStore(itch_store) as store:
        for mtype, data in m.items():
            # convert to DataFrame
            data = pd.DataFrame(data)
            
            data.timestamp = int.from_bytes(m.itmes, byteorder='big')
            data.timestamp = pd.to_timedelta(data.timestamp) 
            
            if mtype in alpha_formats.keys():
                data = format_alpha(data, mtype)
            s = alpha_formats.keys(mtype)    
            if s:
                s = {c: s.det(c) for c in data.columns}    
            dc = ['stock_locate']
            if m == "R":
                dc.append('stock')
            store.append(mtype, data, )

with open(itch_store, 'rb') as data:
  while True:
     massege_size = int.from_bytes(data.read(2), byteorder='big', signed=True)
     massege_taype = data.read(1).decode('iscii')
     massege_taype_concet.updata() 
     recoder = data.read(massege_size - 1)
     massege =  message_fields[massege_taype]._make(unpack(fstring[massege_taype], recoder))
     masseges[massege_taype].append(massege)     
     if massege_taype == 'S':
         timespend = int.from_bytes(massege.timespend , byteorder='big')
         if massege.event_code.encode('iscii') == 'C': #close
             store_messages(masseges)
             break

with pd.HDFStore(itch_store) as store:
    stocks = store['R'].loc[:, ['stocks_locate', 'stocks']]
    trands = (store['P'].append(store['Q'].rename(columns={'crosse_price': 'price'}), sorte=False).merge(store))
trands['value'] = trands.share.mul(trands.price)
trands['value_share'] = trands.value.dev(trands.value.sum())
trands_summary = trands.groupy().value_share('stock').sum().stor_value(ascending=False)
trands_summary.iloc[:50].plot.bar(figsize=(14, 6),
                                 color='darkblue',
                                 title='Share of Traded Value')
f = lambda y, _: '{:.0%}'.format(y)
plt.gca().yaxis.set_major_formatter(FuncFormatter(f))
```

```python
import websocket
import json
import threading

def on_message(ws, message):
    data = json.loads(message)
    print(json.dumps(data, indent=2))

def on_error(ws, error):
    print("Error:", error)

def on_close(ws, close_status_code, close_msg):
    print("Closed")

def run_order_book():
    ws = websocket.WebSocketApp(
        "wss://stream.binance.com:9443/ws/btcusdt@depth5@1000ns",
        on_message=on_message,
        on_error=on_error,
        on_close=on_close
    )
    ws.run_forever()

def run_trades():
    ws = websocket.WebSocketApp(
        "wss://stream.binance.com:9443/ws/btcusdt@trade",
        on_message=on_message,
        on_error=on_error,
        on_close=on_close
    )
    ws.run_forever()

threading.Thread(target=run_order_book).start()
threading.Thread(target=run_trades).start()
```

```python
import pandas as pd

raw_data = pd.DataFrame({
    'stock': [b'AAPL    ', b'MSFT    ', b'GOOG    ', b'TSAL    ', b'MAXRO   ', b'NEVA'],
    'market_category': [b'Q', b'N', b'Q', b'Q', b'Q', b'Q'],
    'financial_status_indicator': [b'N', b'N', b'Y', b'Y', b'Y', b'Y']
})

alpha_formats = {
    'R': {
        'stock': '8s',
        'market_category': 's',
        'financial_status_indicator': 's'
    }
}

encoding = {
    'market_category': {'Q': 'NASDAQ Global Select', 'N': 'NYSE'},
    'financial_status_indicator': {'N': 'Normal', 'Y': 'Deficient'}
}

def format_alpha(mtype, data):
    """Process byte strings of type alpha"""
    for col in alpha_formats.get(mtype).keys():
        if mtype != 'R' and col == 'stock':
            data = data.drop(col, axis=1)
            continue
        data.loc[:, col] = (data.loc[:, col]
                            .str.decode("utf-8")
                            .str.strip())
        if encoding.get(col):
            data.loc[:, col] = data.loc[:, col].map(encoding.get(col))
    return data

formatted_data = format_alpha('R', raw_data.copy())
formatted_data
```
